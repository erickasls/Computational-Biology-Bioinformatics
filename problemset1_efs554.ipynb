###  A
Write a function to calculate the %GC content for a given DNA sequence dna_seq. Make use of string manipulations to do this in a simple manner.

```python
def calculate_gc_content(dna_seq):
    '''
    inputs: dna_seq - string, DNA sequence of interest
    
    outputs: float of %gc content of dna_seq
    
    Calculate the GC content of a given dna sequence
    '''
    #insert your code here
    dna_seq = dna_seq.upper() #add g's and c's and divide it by len dna and X 100; make sure to 'round'
    gc=(dna_seq.count('G')+dna_seq.count('C'))/len(dna_seq)*100
    gc = round (gc,2)
    print("%GC:",gc)
    return gc
```

### B
Transcribe DNA sequence dna_seq to mRNA, assuming dna_seq is the coding sequence. Write two functions to :
* Transcibe dna_seq to the equivalent pre-mRNA sequence.
* Using the coding sequence intervals provided, convert the pre-mRNA sequence to spliced mRNA sequence

```python
def transcribe(dna_seq):
    '''
        inputs: dna_seq, string, the dna sequence of interest
        
        outputs: pre_mrna, string, the transcibed pre-mRNA sequence
        
        transcribe a sequence of dna to pre-mRNA
    '''
    pre_mrna = ''   
    # insert your code here
    pre_mrna = dna_seq.replace("T","U") #Use .replace and call the pre_mrna
    print('pre-mRNA sequence: ',pre_mrna)
    return pre_mrna

def splice(pre_mrna, cds):
    '''
        inputs: pre_mrna, string, the transcribed pre-mRNA sequence to be spliced
                cds: a list of lists that denote the start and stop sites of each codon
                
        outpus: spliced_mrna, string, mRNA that has been spliced according to the coding sequences
                specified in cds
        given all the coding sequences for a piece of pre_mRNA, returned the spliced m_rna
    '''
    spliced_mrna = ''
    # insert your code here
    for n in range(len(cds)):  #use the formula above and you splice with [start:end]
          start=cds[n][0]  
          end=cds[n][1]+1  
          spliced_mrna += pre_mrna [start:end]
    print("spliced mRNA sequence:",spliced_mrna)
    return spliced_mrna
```
